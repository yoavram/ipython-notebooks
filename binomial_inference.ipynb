```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats
```

```python
n = 500
p = 0.2
z = 1.96
```

```python
x = np.random.binomial(n, p)
phat = x/n
ci = z * np.sqrt(phat * (1-phat) / n)
print('{} +- {}'.format(phat, ci))

x = np.random.binomial(n, p, size=10000)

plt.hist(x/n, bins=100, color='k', alpha=0.7)
plt.axvline(phat, color='r')
plt.axvline(phat-ci, color='r', ls='dashed')
plt.axvline(phat+ci, color='r', ls='dashed')
plt.axvline(p, color='b');
```
